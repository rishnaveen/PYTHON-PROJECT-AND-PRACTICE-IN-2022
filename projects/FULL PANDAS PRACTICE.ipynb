# Data filtering using condition

```python
import pandas as pd
df=pd.read_csv('C:\\Users\\Win\\Desktop\\naveenedited.csv')
d=pd.DataFrame(df)
print(d)
```

```python
df
```

```python
df.loc[df["math score"]<70]
```

```python
df.loc[(df["reading score"]<70) & (df["math score"]<70)][0:20]
```

```python
df.loc[(df["reading score"]<70) | (df["math score"]<70)][0:20]
#if anyone condition is true it will return true
```

```python
df.loc[(df["reading score"]<70) & (df["math score"]<70) & (df["writing score"]<60)][0:20]
#this is multiple condition
```

# Conditional changes in pandas

```python
import pandas as pd
df=pd.read_csv('C:\\Users\\Win\\Desktop\\naveenedited.csv')
d=pd.DataFrame(df)
print(d)
```

```python
df
```

```python
df['percentage']=(df['Total score']/500)*100
```

```python
df
```

```python
df["Grade"]="None"
```

```python
df
```

```python
"""
< 40 - Fail
>40 & <60 - Pass
>60 & <70 - FIRST CLASS

"""
```

```python
df.loc[df["percentage"]<40,["Grade"]]="FAIL"
```

```python
df
```

```python
df.loc[(df["percentage"]>40) & (df["percentage"]>60),["Grade"]]="PASS"
```
