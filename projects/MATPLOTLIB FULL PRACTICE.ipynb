```python
#stacked bar chart stacked means palacing one another above
from matplotlib import pyplot as plt
years=["2017","2018","2019","2020",]
placements=[150,275,324,295]

cse=[75,80,47,82]
it=[46,58,35,52]
ece=[67,59,72,62]
ece_start=[cse[i]+it[i]for i in range(len(cse))]

plt.title("PLACEMENT COMPARISION USING STACKED BAR GRAPH",color="red")
plt.bar(years, placements,width=0.4)
plt.bar(years,ece,width=0.4,color="yellow",label="ECE",)
plt.bar(years,it,width=0.4,color="pink",label="IT")
plt.bar(years,cse,bottom=cse,width=0.4,color="black",label="CSE")

plt.ylim(0,350)
plt.xlabel("YEARS",color="green",size=15)
plt.ylabel("PLACEMENTS",color="green",size=15)
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt

student_performance=["Excellent","Good","Average","Poor"]
student_values=[15,25,12,8]

plt.figure(figsize=(8,10))
plt.pie(student_values,labels=student_performance,startangle=0,explode=[0.2,0,0,0],shadow=True,colors=["green","blue","yellow","red"],autopct="%2.1f%%")

plt.legend(title="performances")
plt.show()
```

```python
# histogram chart in matplotlib 
from matplotlib import pyplot as plt
marks=[90,50,40,60,65,85,45,95,60,65]
grade_intervals=[0,50,70,100]
plt.title("STUDENT GRADE",color="blue",size=15)
plt.hist(marks,grade_intervals,histtype="stepfilled",rwidth=0.7,facecolor="k")
plt.xticks([0,50,70,100])
plt.xlabel("PERCENTAGE",color="blue",size=15)
plt.ylabel("NO OF STUDDENT",color="blue",size=15)
plt.show()
```
